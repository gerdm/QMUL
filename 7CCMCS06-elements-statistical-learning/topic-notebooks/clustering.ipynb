# Hierarchical Clustering

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import AgglomerativeClustering
```

```python
%config InlineBackend.figure_format = "retina"
```

```python
X, _ = make_moons(300, noise=0.02, random_state=314)
plt.scatter(*X.T)
```

```python
single_linkage = AgglomerativeClustering(n_clusters=2, linkage="single")
complete_linkage = AgglomerativeClustering(n_clusters=2, linkage="complete")
```

```python
complete = complete_linkage.fit_predict(X)
single = single_linkage.fit_predict(X)
```

```python
c_complete =["tab:orange" if v else "tab:blue" for v in complete]
c_single = ["tab:orange" if v else "tab:blue" for v in single]
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 4))

labels = ["Single", "Complete"]
colors = [c_single, c_complete]

for label, colors, axi in zip(labels, colors, ax):
    axi.scatter(*X.T, c=colors)
    axi.set_title(f"{label} Linkage", fontsize=14)
```
