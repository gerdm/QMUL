```python
import numpy as np
import matplotlib.pyplot as plt
from fractions import Fraction
```

```python
%config InlineBackend.figure_format = "retina"
plt.rcParams["axes.spines.right"] = False
plt.rcParams["axes.spines.top"] = False
```

```python
def plot_map(f, xmin, xmax, range_size=200):
    """
    f: function
        The iterating map
    x0: Fraction, float
        Initial value condition
    n_iterations: None, int
        Number of iterations over the map
    """
    xrange = np.linspace(xmin, xmax, range_size)
    plt.plot(xrange, f(xrange))
    plt.plot(xrange, xrange, c="tab:gray", alpha=0.4, linestyle="--")
    plt.ylim(0, 1)
    plt.grid(alpha=0.6)
    plt.xlabel("$x_n$", fontsize=13)
    plt.ylabel("$x_{n+1}$", fontsize=13)
    plt.title("$T(x)$", fontsize=15);
```

## Q1. Three maps

```python
def f1(x): return x / 2 + 1 / 4

plot_map(f1, 0, 1)
plt.axvline(x=0.5, linestyle="dashed", color="black");
```

```python
xvals = np.array([0, 0.5])
```

```python
colors = ["tab:blue", "tab:red"]
plt.axhline(y=0, c="tab:gray", zorder=1)
plt.axhline(y=-0.2, c="tab:gray", zorder=1)


plt.scatter(xvals, [0, 0], zorder=2, color=colors)
plt.scatter(f1(xvals), [-0.2, -0.2], zorder=2, color=colors)

plt.plot([xvals[0], f1(xvals[0])], [0, -0.2], linestyle="--")
plt.plot([xvals[1], f1(xvals[1])], [0, -0.2], linestyle="--")

plt.grid()
```

### $f_B(x)$

```python
def fB(x): return 4 * x * (1 - x)

plot_map(fB, 0, 1)
plt.axvline(x=0.5, linestyle="dashed", color="black");
```

```python
I0 = np.array([0, 1/2])
fB(I0)
```

```python
I1 = np.array([1/2, 1])
fB(I1)
```

### $f_C(x)$

```python
def fC(x): return (1 - 2 * x) * (0 <= x) * (x <= 1/2) + 2 * (x - 1/2) / 3 * (1/2 < x) * (x <= 1)

plot_map(fC, 0, 1)
plt.axvline(x=0.5, linestyle="dashed", color="black");
```

```python
I0 = np.array([0, 1/2])
fC(I0)
```

```python
I1 = np.array([1/2, 1])
fC(I1)
```
