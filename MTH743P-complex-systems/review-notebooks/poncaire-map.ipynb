# Poncaire map

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%config InlineBackend.figure_format = "retina"
```

```python
X = np.mgrid[-2:2:0.1, -3:3:0.1][::-1]
```

```python
def f(x, y, μ):
    ẋ = y + μ * x - x * (x ** 2 + y ** 2)
    ẏ = -x + μ * y - y * (x ** 2 + y ** 2)
    
    return ẋ, ẏ
```

```python
mu = 1

Ẋ = f(*X, mu)
plt.streamplot(*X, *Ẋ)
plt.axis("equal");
```
