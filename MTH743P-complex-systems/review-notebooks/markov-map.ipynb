# Markov Maps and Markov Partitions

Def

A map $f: I\to I$ is called a Markov map if there exists a partition $\{I_0, I_1, \ldots, I_{N-1}\}$ (called a Markov partition) if, for every $k,l \in \{0, \ldots, N-1\}$

1. Either $f(\text{int}(I_k)) \cap \text{int}(I_l) = \{\}$
2. or $\text{int}(I_l) \subseteq f(\text{int}(I_k))$

```python
import numpy as np
import matplotlib.pyplot as plt
from fractions import Fraction
```

```python
%config InlineBackend.figure_format = "retina"
plt.rcParams["axes.spines.right"] = False
plt.rcParams["axes.spines.top"] = False
```

```python
def plot_map(f, xmin, xmax, range_size=200):
    """
    f: function
        The iterating map
    x0: Fraction, float
        Initial value condition
    n_iterations: None, int
        Number of iterations over the map
    """
    xrange = np.linspace(xmin, xmax, range_size)
    plt.plot(xrange, f(xrange))
    plt.plot(xrange, xrange, c="tab:gray", alpha=0.4, linestyle="--")
    plt.axis("equal")
    plt.ylim(0, 1)
    plt.grid(alpha=0.6)
    plt.xlabel("$x_n$", fontsize=13)
    plt.ylabel("$x_{n+1}$", fontsize=13)
    plt.axvline(x=0.5, linestyle="dotted", c="black")
    plt.title("$T(x)$", fontsize=15);
```

### Tent Map

$$
    T(x) = \begin{cases}
        2 x & 0 \leq x \leq 1/2 \\
        2 - 2x & 1/2 < x \leq 1
    \end{cases}
$$

```python
def T(x):
#     val = 2 * x * (0 <= x) * (x <= 1/2) + 2 * (1 - x) * (1/2 < x) * (x <= 1)
    val = 1 - np.abs(2 * x - 1)
    return val
```

```python
plot_map(T, 0, 1)
```

```python
x = Fraction(4999, 10000)
#x = Fraction(2, 5)

map_history = [x]
for _ in range(100):
    x = T(x)
    map_history.append(x)

plt.figure(figsize=(10, 4))
plt.plot(map_history, marker="o")
```

```python
def T(x):
    val = (1 - 2 * x) * (0 <= x) * (x <= 1/2) + (x - 1/2) * (1/2 < x) * (x <= 1)
    return val
```

```python
plot_map(T, 0, 1)
```
