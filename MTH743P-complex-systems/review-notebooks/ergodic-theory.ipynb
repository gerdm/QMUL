# Ergodic Theory

```julia
using Plots
```

```julia
# Logistic map function
f(x, r) = r * x * (1 - x)
# invariant measure for the logistic(r=4) map
ρ(x) = 1 / (π * sqrt(x * (1 - x)))

r = 3.9999
nits = 2000
xn = 0.5
xvals = Array{Float64, 1}()

for it=1:nits
    xn = f(xn, r)
    append!(xvals, xn)
end
```

```julia
xrange = 0.01:0.01:0.999

l = @layout [a b]
p1 = scatter(xvals, label=nothing, ylabel="xn")
p2 = histogram(xvals, bins=50, label="histogram", ylabel="ρ(x)", normalized=true)
plot!(xrange, ρ.(xrange), linewidth=4, label="ρ(x)")
plot(p1, p2, layout=l, size=(800, 300))
```

$\rho(x) = \frac{1}{\pi\sqrt{x(1-x)}}$ is an invariant density measure of the logistic (r=4) map
