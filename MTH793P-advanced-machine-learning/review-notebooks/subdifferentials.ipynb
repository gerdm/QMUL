```python
def f(x): return x ** 2
def fprime(x): return 2 * x
```

## First order condition for convex functions

A function $f$ with convex domain is convex iff

$$
    f(y) \geq f(x) + \nabla f(x)^T (x-y) \ \forall x,y\in \text{dom}(f)
$$

Hence,

$$
    f(y) - f(x) \geq + \nabla f(x)^T (x-y) \ \forall x,y\in \text{dom}(f)
$$

```python
x = 1.2
dfx = fprime(x)
xrange = np.linspace(-2, 2.2, 100)
plt.plot(xrange, f(xrange), linewidth=2, label=r"$f(y)$")
plt.plot(xrange, f(x) + fprime(x) * (xrange - x), linewidth=2, label=r"$f(x) + \nabla f(x) (x-y)$")
plt.legend(fontsize=13)
plt.ylim(-0.5, 4);
```
