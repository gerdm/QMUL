# The Logistic Equation

The logistic equation is given by

$$
    \dot N = r N\left(1 - \frac{N}{k}\right)
$$

This ODE can be solved analytically

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%config InlineBackend.figure_format = "retina"
```

```python
class Logistic:
    def __init__(self, N0, k, r):
        self.N0 = N0
        self.k = k
        self.r = r
    
    def N(self, t):
        num = self.k * self.N0
        den = self.N0 - (self.N0 - self.k) * np.exp(-self.r * t)
        return num / den
```

```python
t = np.linspace(0, 3, 100)

k_vals = [99, 99.5, 100, 100.5, 101]

for k in k_vals:
    logistic = Logistic(100, k, 1)
    plt.plot(t, logistic.N(t), label=f"k={k}")
plt.legend()
plt.title(r"$\dot N = rN\left(1 - \frac{N}{k}\right)$", fontsize=15)
plt.xlabel("time", fontsize=13)
plt.ylabel(r"$N(t)$", fontsize=13)
plt.grid();
```
