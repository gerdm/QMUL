# Conservative and Gradient Systems

$$
    \frac{\partial E}{\partial t} = -\dot{\bf x}_g \cdot \dot{\bf x}_c = 0
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%config InlineBackend.figure_format = "retina"
```

```python
def f(x):
    x, y = x
    return  y, x - x ** 3
    
vmin, vmax, step = -1.5, 1.5 + 0.5, 0.5
X = np.mgrid[-1:1.5:step, vmin:vmax:step][::-1]
X_dot = np.apply_along_axis(f, 0, X)
plt.streamplot(*X, *X_dot, density=1.1, color="tab:gray")
```

### An example

```python
def f(x):
    x, y = x
    return np.sin(y), x * np.cos(y)
    
vmin, vmax, step = -np.pi, np.pi + 0.5, 0.1
X = np.mgrid[vmin:vmax:step, vmin:vmax:step][::-1]
X_dot = np.apply_along_axis(f, 0, X)
```

```python
F = np.apply_along_axis(lambda x: -x[0] * np.sin(x[1]), 0, X)

plt.figure(figsize=(6, 6))
plt.streamplot(*X, *X_dot, density=1.1, color="tab:blue")
plt.contour(*X, F, cmap="bone")
```

```python
def f(x):
    x, y = x
    return y - x * (x ** 2 + y ** 2), -x - y * (x ** 2 + y ** 2)
    
vmin, vmax, step = -1, 1 + 0.5, 0.1
X = np.mgrid[vmin:vmax:step, vmin:vmax:step][::-1]
X_dot = np.apply_along_axis(f, 0, X)
```

```python
plt.figure(figsize=(6, 6))
plt.streamplot(*X, *X_dot, density=1, color="tab:blue")
```
