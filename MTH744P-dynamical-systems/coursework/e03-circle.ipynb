## 4.1.7

$$
    \dot\theta = \sin(k\theta); k\in\mathbb{N}
$$

```python
def f(theta, k): return np.sin(k * theta) 
```

```python
np.sin(k * np.pi / k)
```

```python
plt.plot(Î¸_range, f(Î¸_range, k))

plt.axhline(y=0, c="black", linestyle="--")
plt.xticks(xtick_vals, xticks);

plt.axvline(x=3 * np.pi / 4, c="tab:orange")
plt.axvline(x=7 * np.pi / 4, c="tab:orange")
plt.grid(alpha=0.7)
```

```python
k = 3
Î¸_eval = np.arange(0, 2 * np.pi, np.pi / (k * 3))
dynamics.plot_circular_dynamics(Î¸_eval, lambda theta: f(theta, k), rtol=0.2)
plt.title(r"$\dot\theta = \sin(k\theta)$")
```
