```python
Î¸_range = np.linspace(0, 2 * np.pi, 100)
def f(x): return 1 + 2 * np.cos(x)
```

```python
xtick_vals = [0, np.pi / 2, np.pi, 3 * np.pi / 2, 2 * np.pi]
xticks = [r"$0$", r"$\pi/2$", r"$\pi$", r"$3\pi/2$", r"$2\pi$"]
```

```python
Î¸_eval = np.arange(0, 2 * np.pi, np.pi / (2 * 3))
dynamics.plot_circular_dynamics(Î¸_eval, f, rtol=0.1)
```

```python
plt.plot(Î¸_range, f(Î¸_range))

plt.axhline(y=0, c="black", linestyle="--")
plt.xticks(xtick_vals, xticks);

plt.axvline(x=2 * np.pi / 3, c="tab:orange")
plt.axvline(x=4 * np.pi / 3, c="tab:orange")
plt.grid(alpha=0.7)
```
