# Noninformative priors

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import seed
from scipy.stats import norm
```

```python
%config InlineBackend.figure_format = "retina"
```

```python
def p̃(x, N, σ2):
    return np.exp(-x ** 2 / (2 * σ2)) / np.sqrt(σ2) ** N
```

> The idea that a uniform prior can be used to represent ignorance is non self-consistent

In the following graph we plot the difference between a uniform prior for the variance and the equivalent (non-uniform) prior for the precision.

```python
sigma_range = np.linspace(1e-4, 1, 200)

seed(314)
n_samp = 10
Norm = norm(loc=0, scale=0.4)
x = N.rvs(n_samp).sum()

fig, ax = plt.subplots()
ax2 = ax.twinx()
p1 = ax.plot(sigma_range, p̃(x, n_samp, sigma_range), label="variance-prior")
p2 = ax2.plot(sigma_range, p̃(x, n_samp, sigma_range) * 1 / (1 + sigma_range), c="tab:orange", label="precision-prior")

ax.set_ylabel(r"$p(\sigma^2)$", fontsize=13)
ax2.set_ylabel(r"$p(\tau)$", fontsize=13)

ax.axvline(x=0, linestyle="--", c="tab:gray")
ax.axvline(x=1, linestyle="--", c="tab:gray")

legends = p1 + p2
labels = [l.get_label() for l in legends]

plt.legend(legends, labels);
```
